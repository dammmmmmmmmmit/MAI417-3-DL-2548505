```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Dropout
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

print("TensorFlow version :", tf.__version__)
print("NumPy version      :", np.__version__)
print("All libraries imported successfully!")
```

```python
with open("training_text.txt", "r") as f:
    raw_text = f.read()

# Lowercase normalizes vocabulary (fewer unique chars to learn)
text = raw_text.lower()

print("=" * 45)
print("         TEXT STATISTICS")
print("=" * 45)
print(f"  Total characters : {len(text)}")
print(f"  Total lines      : {len(text.splitlines())}")
print(f"  Unique characters: {len(set(text))}")
print("=" * 45)
print("\nFirst 300 characters of loaded text:\n")
print(text[:300])
```

Text preprocessing begins with loading raw data and normalizing it. Converting to lowercase reduces vocabulary size by merging uppercase and lowercase versions of the same letter (e.g., 'T' and 't' become one token).
What's happening: The text file is read, converted entirely to lowercase, and basic statistics (total characters, total lines, sample preview) are printed for inspection.

```python
# ============================================================
# BLOCK 4: Create Character Vocabulary and Vectorize
# ============================================================

# Sorted set of all unique characters = our vocabulary
vocab      = sorted(set(text))
vocab_size = len(vocab)

# Bidirectional lookup tables
char_to_idx = {char: idx for idx, char in enumerate(vocab)}
idx_to_char = {idx: char for idx, char in enumerate(vocab)}

print("=" * 45)
print("         VOCABULARY INFO")
print("=" * 45)
print(f"  Vocabulary size  : {vocab_size}")
print(f"  Characters       : {''.join(vocab)}")
print("=" * 45)
print("\nSample char → index mappings:")
for char in list(vocab)[:8]:
    print(f"   '{char}'  →  {char_to_idx[char]}")

print("\nSample index → char mappings:")
for idx in range(8):
    print(f"   {idx}  →  '{idx_to_char[idx]}'")
```

Concept: Character-level tokenization maps every unique character to an integer index. This is the "vocabulary" the model only knows characters that appeared in the training text, and everything is expressed as numeric indices.

What's happening: All unique characters in the text are sorted into a vocabulary list. Two lookup dictionaries are built char_to_idx for encoding characters to numbers, and idx_to_char for decoding numbers back to characters during text generation.

```python
# ============================================================
# BLOCK 5: Create Training Sequences
# ============================================================

SEQ_LENGTH = 30   # Characters per input window
STEP       = 2   # Slide window by 3 chars each time

sequences  = []   # Will hold input character windows
next_chars = []   # Will hold the target character for each window

# Sliding window over the full text
for i in range(0, len(text) - SEQ_LENGTH, STEP):
    sequences.append(text[i : i + SEQ_LENGTH])
    next_chars.append(text[i + SEQ_LENGTH])

n_sequences = len(sequences)

print(f"Sequence length   : {SEQ_LENGTH}")
print(f"Step size         : {STEP}")
print(f"Total sequences   : {n_sequences}")
print(f"\nExample INPUT  : '{sequences[0]}'")
print(f"Example TARGET : '{next_chars[0]}'")

# ---- One-Hot Encoding ----
# X shape: (n_sequences, SEQ_LENGTH, vocab_size)
# y shape: (n_sequences, vocab_size)
print("\nOne-hot encoding sequences (please wait)...")

X = np.zeros((n_sequences, SEQ_LENGTH, vocab_size), dtype=np.float32)
y = np.zeros((n_sequences, vocab_size),              dtype=np.float32)

for i, seq in enumerate(sequences):
    for t, char in enumerate(seq):
        X[i, t, char_to_idx[char]] = 1.0          # encode each input char
    y[i, char_to_idx[next_chars[i]]] = 1.0         # encode the target char

print(f"\nX shape : {X.shape}  → (samples, seq_length, vocab_size)")
print(f"y shape : {y.shape}  → (samples, vocab_size)")
print(f"Memory  : {X.nbytes / 1e6:.2f} MB for X")
```

Concept: The sliding window technique creates input-output pairs by moving a fixed-length window across the text one step at a time. The window contents are the input (X) and the very next character after the window is the target (y).

What's happening: Every sequence of SEQ_LENGTH=40 characters becomes one training sample X, and the 41st character becomes the label y. Both are one-hot encoded into binary arrays so the neural network can process them numerically.

```python
# ============================================================
# BLOCK 6: Build the RNN Model
# ============================================================

def build_rnn_model(seq_length, vocab_size):
    model = Sequential([
        SimpleRNN(
            units=128,                              # 128 hidden units
            input_shape=(seq_length, vocab_size),  # (timesteps, features)
            activation='tanh'
        ),Dropout(0.3),      # ← KEY FIX: regularization
        Dense(vocab_size, activation='softmax')    # output: prob over vocab
    ], name="RNN_Model")

    model.compile(
        loss='categorical_crossentropy',  # multi-class classification loss
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

rnn_model = build_rnn_model(SEQ_LENGTH, vocab_size)
rnn_model.summary()
```

Concept: SimpleRNN is the most basic recurrent unit it maintains a single hidden state that gets updated at each time step using the formula ht​=tanh(Wh​ x ht−1​+Wx​ x xt​). It struggles to learn long-range dependencies because gradients can vanish over many timesteps.

What's happening: A Sequential model is defined with one SimpleRNN layer (128 units) that reads the 40-character input sequence and outputs a single hidden state, which is then passed to a Dense layer with softmax activation to produce a probability distribution over all characters.

```python
# ============================================================
# BLOCK 7: Build the LSTM Model
# ============================================================

def build_lstm_model(seq_length, vocab_size):
    model = Sequential([
        LSTM(
            units=128,                              # 128 memory cells
            input_shape=(seq_length, vocab_size),
        ),Dropout(0.3),
        Dense(vocab_size, activation='softmax')    # output: prob over vocab
    ], name="LSTM_Model")

    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

lstm_model = build_lstm_model(SEQ_LENGTH, vocab_size)
lstm_model.summary()
```

Concept: LSTM (Long Short-Term Memory) extends SimpleRNN with three learnable gates forget gate, input gate, and output gate plus a separate cell state Ct​. These gates allow the model to selectively remember or forget information across many timesteps, overcoming the vanishing gradient problem.

What's happening: The architecture is identical to the RNN model but with the SimpleRNN layer replaced by an LSTM layer. Because LSTM maintains both a hidden state and a cell state, it captures longer dependencies in the text, which typically leads to more coherent text generation.

```python
# ============================================================
# FIXED BLOCK 8: Train Both Models (Validation-Monitored)
# ============================================================

EPOCHS     = 80    # Higher ceiling — EarlyStopping will cut it short
BATCH_SIZE = 64    # Smaller batch = noisier gradients = less overfitting

early_stop_rnn = EarlyStopping(
    monitor='val_loss',          # ← KEY FIX: watch validation, not training
    patience=8,
    restore_best_weights=True,
    verbose=1
)
early_stop_lstm = EarlyStopping(
    monitor='val_loss',
    patience=8,
    restore_best_weights=True,
    verbose=1
)

# ---- Train RNN ----
print("=" * 50)
print("  Training RNN Model ...")
print("=" * 50)
rnn_history = rnn_model.fit(
    X, y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.1,        # ← KEY FIX: hold out 10% for validation
    callbacks=[early_stop_rnn],
    verbose=1
)

# ---- Train LSTM ----
print("\n" + "=" * 50)
print("  Training LSTM Model ...")
print("=" * 50)
lstm_history = lstm_model.fit(
    X, y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.1,
    callbacks=[early_stop_lstm],
    verbose=1
)

# ---- Plot Loss (Train vs Validation) ----
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(rnn_history.history['loss'],     label='Train Loss', color='royalblue')
axes[0].plot(rnn_history.history['val_loss'], label='Val Loss',   color='red', linestyle='--')
axes[0].set_title('RNN: Train vs Validation Loss')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(lstm_history.history['loss'],     label='Train Loss', color='darkorange')
axes[1].plot(lstm_history.history['val_loss'], label='Val Loss',   color='red', linestyle='--')
axes[1].set_title('LSTM: Train vs Validation Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ---- Summary ----
print("\n" + "=" * 55)
print("  TRAINING RESULTS SUMMARY")
print("=" * 55)
print(f"  RNN  → Train Loss: {rnn_history.history['loss'][-1]:.4f}"
      f"  |  Val Loss: {rnn_history.history['val_loss'][-1]:.4f}"
      f"  |  Epochs: {len(rnn_history.history['loss'])}")
print(f"  LSTM → Train Loss: {lstm_history.history['loss'][-1]:.4f}"
      f"  |  Val Loss: {lstm_history.history['val_loss'][-1]:.4f}"
      f"  |  Epochs: {len(lstm_history.history['loss'])}")
print()
print("  Healthy training signs:")
print("  ✔ Train loss and Val loss should be close to each other")
print("  ✔ Neither should be near 0.00 (that's memorization)")
print("  ✔ RNN accuracy should now be 70-85%, NOT 97%+")
```

Concept: Both models are trained using supervised learning where categorical cross-entropy loss measures how different the model's predicted character probability distribution is from the true one-hot label. EarlyStopping monitors training loss and halts training if it stops improving, preventing wasted time and overfitting.

What's happening: Each model is fitted on the X and y arrays for up to 50 epochs with a batch size of 128. Training histories are saved so the loss curves of RNN vs LSTM can be plotted side by side for comparison.

```python
# ============================================================
# BLOCK 9: Text Generation Function
# ============================================================

def sample_with_temperature(predictions, temperature=1.0):
    """
    Rescales the prediction distribution using temperature,
    then randomly samples one character index from it.
    """
    predictions = np.asarray(predictions).astype("float64")
    # Apply temperature scaling in log-space for numerical stability
    predictions = np.log(predictions + 1e-10) / temperature
    exp_preds   = np.exp(predictions)
    predictions = exp_preds / np.sum(exp_preds)
    return np.random.choice(len(predictions), p=predictions)


def generate_text(model, seed_text, char_to_idx, idx_to_char,
                  seq_length, num_chars=300, temperature=0.8):
    """
    Generates new text character-by-character using a trained model.

    Args:
        model       : Trained RNN or LSTM model
        seed_text   : Starting string to kick off generation
        char_to_idx : char → index lookup
        idx_to_char : index → char lookup
        seq_length  : Input window size the model was trained on
        num_chars   : Number of new characters to generate
        temperature : Sampling temperature (0.5=focused, 1.5=creative)

    Returns:
        generated_text : String of newly generated characters only
    """
    seed_text = seed_text.lower()

    # Ensure seed is exactly seq_length characters
    if len(seed_text) < seq_length:
        # Left-pad with spaces so seed ends at the right side of the window
        seed_text = " " * (seq_length - len(seed_text)) + seed_text
    else:
        seed_text = seed_text[-seq_length:]

    generated = seed_text  # running buffer

    for _ in range(num_chars):
        # Build one-hot input from the last seq_length characters
        x = np.zeros((1, seq_length, len(char_to_idx)), dtype=np.float32)
        for t, char in enumerate(generated[-seq_length:]):
            if char in char_to_idx:
                x[0, t, char_to_idx[char]] = 1.0

        # Predict next character probabilities
        preds    = model.predict(x, verbose=0)[0]
        next_idx = sample_with_temperature(preds, temperature)
        next_char = idx_to_char[next_idx]

        generated += next_char

    # Return only the newly generated part (strip the seed window)
    return generated[seq_length:]


print("Text generation function ready!")
print()
print("Temperature guide:")
print("  0.5  →  repetitive but accurate to training style")
print("  0.8  →  balanced (recommended)")
print("  1.2  →  more creative but less structured")
```

Concept: Temperature-based sampling controls the creativity of generated text. A temperature below 1.0 sharpens the probability distribution (more predictable output), while above 1.0 flattens it (more random output). Mathematically, pi​=∑j​exp(log(pj​)/T)exp(log(pi​)/T)​ where T is the temperature.

What's happening: The function takes a seed string and iteratively predicts one character at a time it one-hot encodes the last 40 characters as input, feeds it to the model, samples the next character from the predicted distribution using temperature scaling, appends it to the result, and repeats until num_chars new characters are generated.
